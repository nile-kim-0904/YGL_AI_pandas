```python
# subpot: 하나의 페이지에 여러개의 그래프 그림
x1 = np.linspace(0.0,2.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1) # subplot(행의 수, 열의 수, 위치)
plt.plot(x1,y1,'yo-')
plt.title('A table of 2 subplot')
plt.ylabel('Damped oscillation')
#print(ax1)
plt.show()

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.title('B table of 2 subplot')
plt.xlabel('time')
plt.ylabel('Undamped')
#print(ax2)
plt.show()
```

```python
# 선 그래프
# 시도별 진출입 인구수
import pandas as pd

df = pd.read_excel("dataset/시도별 전출입 인구수.xlsx",header=0)
df.head(25)
```

```python
# 데이터 전처리 작업
# 1. NaN 처리
# 전출지의 NaN 자료를 수정 -> 이전 자료의 값으로 대체
# NaN 자료를 수정: method = ffill 이전 자료로 대체, bfill 이후 자료로 대체
# value = 값 -> 지정하는 값으로 수정
df.fillna(method="ffill",inplace=True) # NaN 자료를 수정: ffill 이전 자료로 대체
df.head(25)

# 2. 필요한 자료만 추출
# 서울특별시에서 다른 지역으로 전출한 자료만 추출
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
print(df_seoul.head())
# 필요없는 컬럼 삭제: "전출지별"은 같은 자료이므로 삭제
df_seoul = df_seoul.drop('전출지별',axis=1)
df_seoul

# 3. 컬럼명 또는 인덱스 재설정
# "전입지별"을 "전입지"로 변경
df_seoul = df_seoul.rename({"전입지별":"전입지"},axis=1)
df_seoul

# "전입지" 컬럼을 인덱스로 변경
df_seoul.set_index("전입지",inplace=True)
df_seoul
```

```python
df_seoul.T.plot()
```

```python
# 서울에서 경기도로 전입한 인구수를 그래프로 확인
sr_one = df_seoul.loc["경기도"]
sr_one.head(3)
sr_one.index
sr_one.values

sr_jeju = df_seoul.loc["제주특별자치도"]
sr_one
```

```python
plt.figure(figsize=(15,5)) # 그래프 사이즈 지정
plt.plot(sr_one.index,sr_one.values,marker='o',markersize=7,color='g')
plt.xticks(size=10,rotation='vertical') # x축의 데이터를 세로로 회전
plt.title("서울에서 경기도로 이전한 인구 수") # 그래프의 제목
plt.xlabel("연도",size=20) # x축 제목
plt.ylabel("인구 수",size=20) # y축 제목
plt.legend(labels = ["서울 -> 경기도"], loc = "best",fontsize=13)

# annotate 추가
plt.ylim(50000,800000)
plt.annotate('',
             xy=(20,620000), # 화살표 시작점
             xytext=(2,290000), # 화살표 끝점
             xycoords='data',
             arrowprops=dict(arrowstyle='->',color='blue',lw=3))
plt.annotate("인구 이동 증가 (1970 - 1995)",
             xy=(10,370000), # x좌표,y좌표
             rotation=22,
             va="baseline", # center, top, bottom, baseline
             ha="center", # center, left right
             fontsize=14)
plt.show()
```

```python
# 서울에서 제주도로 이동한 인구수만 검색해서 그래프로 나타내기
df = pd.read_excel("dataset/시도별 전출입 인구수.xlsx")
df.fillna(method="ffill",inplace=True)

mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] == '제주특별자치도')
df = df[mask]

df = df.drop('전출지별',axis=1)

df = df.rename({"전입지별":"전입지"},axis=1)

df.set_index("전입지",inplace=True)

df.T.plot()
```

```python
# 서울 -> 경기도
# 서울 -> 제주로 이동한 인구수의 그래프를 동시에 표현


fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

# ax1에 서울 -> 경기도: sr_one
ax1.plot(sr_one,marker='o',color='b',markersize=10, label='서울 -> 경기')
ax2.plot(sr_jeju,marker='^',color='r',markersize=10, label='서울 -> 제주')

ax1.set_title("서울 경기 인구 이동")
ax2.set_title("서울 제주 인구 이동")

ax1.legend(loc='best')
ax2.legend(loc='best')

# y축 범위 (최소값, 최대값)
#ax1.set_ylim(50000,800000)
#ax2.set_ylim(0,20000)

# x축 연도 라벨 지정 -> 45도 기울여서
ax1.set_xticklabels(sr_one.index,rotation=45)
ax2.set_xticklabels(sr_jeju.index,rotation=45)

# 눈금 라벨 크기
ax1.tick_params(axis='x',labelsize=10)
ax2.tick_params(axis='y',labelsize=10)

plt.show()
```

```python
# 남북한 전령 발전량
df = pd.read_excel("dataset/남북한발전전력량.xlsx",convert_float=True)
#df.fillna(method="ffill",inplace=True)

# 북한의 자료만 가져 옴
df = df.loc[5:9]

# '전력량 (억㎾h)' 컬럼을 삭제
#df.columns
df = df.drop('전력량 (억㎾h)',axis='columns')

# 발전 전력별 컬럼을 인덱스로 지정
#df.columns
df = df.set_index('발전 전력별')

df = df.T

# 연도별 증감률, '합계' 컬럼명을 '총 발전량'으로 변경
df = df.rename(columns = {'합계':'총발전량'})

# 총발전량 - 1년 -> 이전 년도의 총발전량
df['총발전량 - 1년'] = df['총발전량'].shift(1)

# 증감률 (총발전량/이전 년도의 발전량 -1) * 100
df['증감률'] = ((df['총발전량']/df['총발전량 - 1년'])-1)*100

# 그래프 그리기: 수력, 화력 -> 증감률
ax1 = df[['수력','화력']].plot(kind='bar',figsize=(20,10),stacked=True)
ax2 = ax1.twinx()
ax2.plot(df.index,df['증감률'],ls='--',marker='o',color='r',lw=3,label='전년대비 증감률(%)')

ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도',size=20)
ax1.set_ylabel('발전량 (억 Kwh)',size=20)
ax2.set_ylabel('전년 대비 증감률 (%)')

plt.title('북한 전력 발전량')
ax1.legend(loc='best')

df.head(10)
```
